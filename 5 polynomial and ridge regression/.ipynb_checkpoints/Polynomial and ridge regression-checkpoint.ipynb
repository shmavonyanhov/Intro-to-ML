```python
#Fitting a polynomial regression
```

```python
from regression import fit_polynomial_regression
```

```python
# Scatter plot area vs price.

plt.figure(figsize=(15,10))
plt.scatter(xTest, yTest, s=3)
plt.xlabel('area in square meters')
plt.ylabel('price in 1000$')
plt.title('Apartment prices in Yerevan Center')

beta = fit_polynomial_regression(xTrain, yTrain, degree=2)
curve_x = np.arange(20, 160, 1)
curve_y = [sum(beta[i] * x ** i for i in range(len(beta))) for x in curve_x]
plt.plot(curve_x, curve_y, color='red')

plt.show()
```

```python
print('degree, rmse_train, rmse_test')
train_errors = []
test_errors = []
for n in range(1, 10):
    beta = fit_polynomial_regression(xTrain, yTrain, degree=n)
    curve_x = np.arange(20, 160, 1)
    curve_y = [sum(beta[i] * x ** i for i in range(len(beta))) for x in curve_x]
    squared_error_train = np.sqrt(sum((y - sum(beta[i] * x ** i for i in range(len(beta)))) ** 2
                        for x, y in zip(xTrain, yTrain)) / len(xTrain))
    squared_error_test = np.sqrt(sum((y - sum(beta[i] * x ** i for i in range(len(beta)))) ** 2
                        for x, y in zip(xTest, yTest)) / len(xTest))
    print(n, squared_error_train.round(2), squared_error_test.round(2))
    train_errors.append(squared_error_train)
    test_errors.append(squared_error_test)
    
plt.plot(range(1, 1 + len(train_errors)), train_errors, label='train err')
plt.xlabel('degree')
plt.ylabel('squared_error_train')


plt.plot(range(1, 1 + len(test_errors)), test_errors, label='test err')
plt.xlabel('degree')
plt.ylabel('squared_error_test')

plt.legend()
plt.show()
```

```python
# Scatter plot area vs price.
plt.clf() 
plt.figure(figsize=(15,10))
plt.scatter(xTest, yTest, s=3, color='k')
plt.xlabel('area in square meters')
plt.ylabel('price in 1000$')
plt.title('Apartment prices in Yerevan Center')

beta = fit_one_dimensional_linear_regression(np.array(xTrain), np.array(yTrain))
curve_x = np.arange(30, 140, 1)
curve_y = [sum(beta[i] * x ** i for i in range(len(beta))) for x in curve_x]
plt.plot(curve_x, curve_y, color='red', label='Linear Regression Line')

beta = fit_polynomial_regression(xTrain, yTrain, degree=5)
poly_curve_y = [sum(beta[i] * x ** i for i in range(len(beta))) for x in curve_x]
plt.plot(curve_x, poly_curve_y, color='green', label='Polynomial Regression Line')


plt.legend()
plt.show()
```

```python
#Fitting Ridge regression
```

```python
from regression import fit_ridge_polynomial_regression
```

```python
print('degree, poly_rmse_test, ridge_rmse_test')
err1_test, err2_test = [], []
err1_train, err2_train = [], []
for n in range(1, 10):
    curve_x = np.arange(20, 160, 1)
    r_errors_test, p_errors_test = [], []
    r_errors_train, p_errors_train = [], []

    beta = fit_ridge_polynomial_regression(xTrain, yTrain, degree=n, l=1.5)
    
    curve_y = [sum(beta[i] * x ** i for i in range(len(beta))) for x in curve_x]
    r_errors_test.append(np.sqrt(sum((y - sum(beta[i] * x ** i for i in range(len(beta)))) ** 2
                      for x, y in zip(xTest, yTest)) / len(xTest)))
    r_errors_train.append(np.sqrt(sum((y - sum(beta[i] * x ** i for i in range(len(beta)))) ** 2
                      for x, y in zip(xTrain, yTrain)) / len(xTrain)))
    
    beta = fit_polynomial_regression(xTrain, yTrain, degree=n)
    
    curve_y = [sum(beta[i] * x ** i for i in range(len(beta))) for x in curve_x]
    
    p_errors_test.append(np.sqrt(sum((y - sum(beta[i] * x ** i for i in range(len(beta)))) ** 2
                    for x, y in zip(xTest, yTest)) / len(xTest)))
    p_errors_train.append(np.sqrt(sum((y - sum(beta[i] * x ** i for i in range(len(beta)))) ** 2
                    for x, y in zip(xTrain, yTrain)) / len(xTrain)))
    
    print('{:6d}  {:10,.1f}   {:,.1f}'.format(n, np.mean(r_errors_test), np.mean(p_errors_test)))
    err1_test.append(np.mean(r_errors_test))
    err2_test.append(np.mean(p_errors_test))
    err1_train.append(np.mean(r_errors_train))
    err2_train.append(np.mean(p_errors_train))
    
plt.plot(range(1, len(err1_test) + 1), err1_test, label='poly')
plt.plot(range(1, len(err2_test) + 1), err2_test, label='ridge poly')
plt.title('Test Errors')
plt.legend()
plt.show()

plt.plot(range(1, len(err1_train) + 1), err1_train, label='poly')
plt.plot(range(1, len(err2_train) + 1), err2_train, label='ridge poly')
plt.title('Train Errors')
plt.legend()
plt.show()
```

```python
#Polynomial regression
```

```python
degree = 5
poly_features = PolynomialFeatures(degree=degree)
#print(type(poly_features))
# transforms the existing features to higher degree features.
X_train_poly = poly_features.fit_transform(np.array(xTrain).reshape(-1, 1))
print(xTrain)
print(X_train_poly[0])
print(X_train_poly.shape)
# fit the transformed features to Linear Regression
poly_model = LinearRegression()
poly_model.fit(X_train_poly, yTrain)

# predicting on training data-set
y_train_predicted = poly_model.predict(X_train_poly)

# predicting on test data-set
y_test_predict = poly_model.predict(poly_features.fit_transform(np.array(xTest).reshape(-1, 1)))

# evaluating the model on training dataset
rmse_train = np.sqrt(mean_squared_error(yTrain, y_train_predicted))

# evaluating the model on test dataset
rmse_test = np.sqrt(mean_squared_error(yTest, y_test_predict))
```

```python
#Ridge regression
```

```python
from sklearn.linear_model import Ridge


degree = 5

poly_features = PolynomialFeatures(degree=degree)
  
# transforms the existing features to higher degree features.
X_train_poly = poly_features.fit_transform(np.array(xTrain).reshape(-1, 1))

clf = Ridge(alpha=1.5)
clf.fit(X_train_poly, yTrain)

# predicting on training data-set
y_train_predicted = clf.predict(X_train_poly)

# predicting on test data-set
y_test_predict = clf.predict(poly_features.fit_transform(np.array(xTest).reshape(-1, 1)))

# evaluating the model on training dataset
rmse_train = np.sqrt(mean_squared_error(yTrain, y_train_predicted))

# evaluating the model on test dataset
rmse_test = np.sqrt(mean_squared_error(yTest, y_test_predict))
```
