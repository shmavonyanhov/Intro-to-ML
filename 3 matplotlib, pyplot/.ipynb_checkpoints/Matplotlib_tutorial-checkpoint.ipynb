## Adding titles and labels 

````python
x = [1, 2, 3]
y = [5, 7, 4]

plt.plot(x, y, linestyle = 'dashed', color = 'red') #add a parameter linestyle='dotted'

plt.xlabel('Numbers')
plt.ylabel('Values')
#plt.title('Our first graph')
plt.title('Our first graph\nSubtitle1')
plt.xlim((1,5))
plt.ylim((1,7)) 
#plt.xlim(left=1)


plt.show()
````

## Legends

````python
x = [1, 2, 3]
y = [5, 7, 4]

x2 = [1, 2, 3]
y2 = [14, 14, 12]

plt.plot(x, y, label = "First Line")
plt.plot(x2, y2, label = "Second Line")

plt.xlabel('Numbers')
plt.ylabel('Values')

plt.legend()
plt.title('Our first graph')
#plt.title('Our first graph\nSubtitle1')

plt.show()
````

## Bar charts and histograms

````python
# bar chart - comparing things
# histogram - shows the distribution
````

````python
x = [1, 2, 3]
y = [5, 7, 4]


plt.bar(x, y)

plt.xlabel('Numbers')
plt.ylabel('Values')


plt.title('Bar Graph')
plt.xticks(x,['A', 'B', 'C'])

plt.show()
````

````python
x = ['G1', 'G2', 'G3', 'G4']  #student groups
y = [5, 7, 4, 6]  #mean grades per group


plt.bar(x, y)

plt.xlabel('Groups')
plt.ylabel('Grades')

plt.title('Compare grades')

plt.show()
````

````python
x = ['G1', 'G2', 'G3', 'G4']  #student groups
y = [5, 7, 4, 6]  #mean grades per group


plt.barh(x, y)

plt.xlabel('Grades')
plt.ylabel('Groups')

plt.title('Compare grades')

plt.show()
````

````python
data = {'men_means':[22, 30, 33, 30, 26],'women_means':[25, 32, 30, 35, 29]}
df = pd.DataFrame(data)
df
````

````python
import numpy as np


# data to plot

men_means = df['men_means']
women_means = df['women_means']


index = np.arange(len(men_means)) #[0, 1, 2, 3, 4]
bar_width = 0.35

plt.bar(index - bar_width, men_means, bar_width, color='c', label='Men')

plt.bar(index, women_means, bar_width, color='r', label='Women') #alpha = 0.5

plt.xlabel('Gender Group')
plt.ylabel('Mean Scores')
plt.title('Score comparison')
plt.xticks(index, ['G1', 'G2', 'G3', 'G4', 'G5'])
plt.legend()


plt.show()
````

## Pie chart

````python
slices = [7, 2, 2, 13]
activities = ['sleeping', 'eating', 'working', 'playing']
cols = ['m', 'c', 'r', 'y']

plt.pie(slices, labels = activities, colors = cols, startangle = 90, shadow = True, explode = (0, 0.1, 0, 0), autopct = "%1.1f%%") 
#add startangle = 90, shadow = True, explode = (0, 0.1, 0, 0), autopct = "%1.1f%%"


plt.title('Pie chart')


plt.show()
````

## Subplots

````python
fig, axs = plt.subplots(2)
fig.tight_layout(h_pad = 5)

x1 = [1, 2, 3, 4]
y1 = [8, 2, 5, 6]

x2 = [1, 2, 3, 4]
y2 = [3, 6, 1, 4]

axs[0].plot(x1, y1)
axs[0].set_title('plot1')
axs[0].set_xlabel('x1')
axs[0].set_ylabel('y1')

axs[1].scatter(x2, y2)
axs[1].set_title('plot2')
axs[1].set_xlabel('x2')
axs[1].set_ylabel('y2')
````

````python
fig, axs = plt.subplots(2, 2)
fig.tight_layout(h_pad = 5) #w_pad = 20

x1 = [1, 2, 3, 4]
y1 = [8, 2, 5, 6]

x2 = [1, 2, 3, 4]
y2 = [3, 6, 1, 4]

axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('plot1')
axs[0, 0].set_xlabel('x1')
axs[0, 0].set_ylabel('y1')

axs[1, 0].scatter(x2, y2)
axs[1, 0].set_title('plot2')
axs[1, 0].set_xlabel('x2')
axs[1, 0].set_ylabel('y2')
````

````python
data = np.random.rand(5, 4)
heatmap = plt.pcolor(data)

for y in range(data.shape[0]):
    for x in range(data.shape[1]):
        plt.text(x + 0.5, y + 0.5, '%.2f' % data[y, x],
                 horizontalalignment='center',
                 verticalalignment='center',
                 )

plt.colorbar(heatmap)

plt.show()
````

````python
data = {'grades':[9, 8, 6, 10, 8, 8, 5, 7, 8, 9, 9, 10, 7, 3, 7], 
          'work':[5, 4, 1, 3, 3, 5, 1, 2, 3, 4, 5, 6, 2, 0, 2]}
df = pd.DataFrame(data)
df
````

````python
import matplotlib.pyplot as plt
print(df.corr())
plt.matshow(df.corr())
plt.show()
````

````python
import matplotlib.pyplot as plt
print(df.corr())

alpha = list(df.corr().index)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)


cax = ax.matshow(df.corr())
fig.colorbar(cax)

ax.set_xticklabels(['']+alpha)
ax.set_yticklabels(['']+alpha)
plt.show()
````
